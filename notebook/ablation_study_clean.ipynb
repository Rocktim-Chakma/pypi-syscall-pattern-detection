```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
DATA_PATH = "data/combined_traces_dataset.csv"
```

```python
df = pd.read_csv(DATA_PATH)
X_full = df.drop(columns=["Package", "Level"]).select_dtypes(include=["number"])
y = df["Level"]

print(X_full.shape)
```

```python
# Full feature model
X_train, X_test, y_train, y_test = train_test_split(
    X_full, y, test_size=0.2, stratify=y, random_state=42
)

rf_full = RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1)
rf_full.fit(X_train, y_train)
pred_full = rf_full.predict(X_test)

print("FULL FEATURE MODEL")
print("Accuracy:", accuracy_score(y_test, pred_full))
print("Precision:", precision_score(y_test, pred_full))
print("Recall:", recall_score(y_test, pred_full))
print("F1:", f1_score(y_test, pred_full))
```

```python
# Without syscall features
syscall_cols = [c for c in X_full.columns if c.startswith("syscall")]
X_no_syscall = X_full.drop(columns=syscall_cols)

X_train, X_test, y_train, y_test = train_test_split(
    X_no_syscall, y, test_size=0.2, stratify=y, random_state=42
)

rf_no_sys = RandomForestClassifier(n_estimators=300, random_state=42, n_jobs=-1)
rf_no_sys.fit(X_train, y_train)
pred_no_sys = rf_no_sys.predict(X_test)

print("WITHOUT SYSCALL FEATURES")
print("Accuracy:", accuracy_score(y_test, pred_no_sys))
print("Precision:", precision_score(y_test, pred_no_sys))
print("Recall:", recall_score(y_test, pred_no_sys))
print("F1:", f1_score(y_test, pred_no_sys))
```

```python
results = pd.DataFrame([
    {
        "Model_Configuration": "Full Behavioral Features",
        "Accuracy": accuracy_score(y_test, pred_full),
        "Precision": precision_score(y_test, pred_full),
        "Recall": recall_score(y_test, pred_full),
        "F1": f1_score(y_test, pred_full),
    },
    {
        "Model_Configuration": "Without Syscall Features",
        "Accuracy": accuracy_score(y_test, pred_no_sys),
        "Precision": precision_score(y_test, pred_no_sys),
        "Recall": recall_score(y_test, pred_no_sys),
        "F1": f1_score(y_test, pred_no_sys),
    }
])

print(results)
```
